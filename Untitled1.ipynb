```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, FancyArrow

# Create figure
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.axis('off')  # Hide axes

def draw_block(x, y, width, height, text):
    ax.add_patch(Rectangle((x, y), width, height, fill=False, linewidth=1.5))
    ax.text(x + width/2, y + height/2, text, ha='center', va='center', fontsize=10, wrap=True)

def draw_arrow(x1, y1, x2, y2):
    ax.add_patch(FancyArrow(x1, y1, x2-x1, y2-y1, width=0.05, length_includes_head=True, head_width=0.3))

# Draw CPU
draw_block(4, 8, 2, 1, "CPU\nAddress Bus\nData Bus\nControl Signals")

# Draw 74LS138 Decoder
draw_block(4, 6, 2, 1, "74LS138\nAddress Decoder")

# Draw Memory Blocks
draw_block(1, 4, 3, 1, "EPROM\n0C00h - 2BFFh")
draw_block(6, 4, 3, 1, "RAM\n2C00h - 3BFFh")

# Draw I/O Blocks
draw_block(1, 1, 3, 1, "Input\n0Ch\n74LS244")
draw_block(4, 1, 2, 1, "Display1\n0Dh\n74LS373\n7-segment")
draw_block(7, 1, 3, 1, "Display2\n0Eh\n74LS373\n7-segment")

# Draw arrows from CPU to Decoder
draw_arrow(5, 8, 5, 7)  # CPU -> Decoder

# Draw arrows from Decoder to EPROM, RAM, I/O
draw_arrow(5, 6, 2.5, 5)  # Decoder -> EPROM
draw_arrow(5, 6, 7.5, 5)  # Decoder -> RAM
draw_arrow(5, 6, 2.5, 2)  # Decoder -> Input
draw_arrow(5, 6, 5, 2)    # Decoder -> Display1
draw_arrow(5, 6, 8, 2)    # Decoder -> Display2

# Draw arrows for data bus (bidirectional)
draw_arrow(5, 7.9, 2.5, 5.9)  # CPU <-> EPROM
draw_arrow(5, 7.9, 7.5, 5.9)  # CPU <-> RAM
draw_arrow(5, 7.9, 2.5, 2.9)  # CPU <-> Input
draw_arrow(5, 7.9, 5, 2.9)    # CPU <-> Display1
draw_arrow(5, 7.9, 8, 2.9)    # CPU <-> Display2

plt.tight_layout()
plt.show()
```
